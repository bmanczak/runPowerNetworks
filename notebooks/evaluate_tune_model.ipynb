```python
class AgentThresholdEnv(BaseAgent):
    def __init__(self, rllib_env, trained_agent, rho_threshold = 0.9):
        BaseAgent.__init__(self, rllib_env.action_space)
        self.trained_agent = trained_agent
        self.rllib_env = rllib_env
        self.steps = 0
        self.rho_threshold = rho_threshold

    def act(self, obs, reward, done):
        
        gym_obs = self.rllib_env.obs_grid2rllib(obs)
        #print(gym_obs)
        #gym_act = self.trained_agent.compute_action(gym_obs)
        #grid2op_act = rllib_env.env_gym.g2op_act
        #grid2op_act = self.rllib_env.env_gym.action_mapper(gym_act, gym_obs)

        if max(obs.rho) < self.rho_threshold:
            gym_act = 0
        else:
            gym_act = self.trained_agent.compute_action(gym_obs)
        
        grid2op_act = self.rllib_env.env_gym.action_mapper(sub_id = gym_act, obs = obs)
        
        return grid2op_act

wrapped_agent = AgentThresholdEnv(rllib_env, agent)
```

```python
from grid2op.Episode import EpisodeData
from tqdm import tqdm 

path_agent = "/Users/blazejmanczak/Desktop/School/Year2/Thesis/runPowerNetworks/runner_log/ppo_5400"
episode_studied = EpisodeData.list_episode(path_agent)
# this_episode = EpisodeData.from_disk(*episode_studied[0])

# this_episode.actions
episode_studied
```

```python
act_distr = defaultdict(int)
c = 0
actions = defaultdict(list)
topo_vects = defaultdict(list)
count_action = defaultdict(int)

for i in tqdm(range(len(episode_studied))):
    this_episode = EpisodeData.from_disk(*episode_studied[i])
    for act, obs in zip(this_episode.actions, this_episode.observations): # all actions in this episode
        topo_vects[i].append(obs.topo_vect)
        if max(obs.rho) < 0.9:
            continue
        else:
           # if len(act.as_dict()) != 0:
            count_action[i] += 1
            actions[i].append(act.as_dict())
            if len(topo_vects[i]) > 1:
                if (topo_vects[i][-1] == topo_vects[i][-2]).all():
                    act_distr[-1] += 1
                else:
                    act_distr[ int(act.as_dict()["set_bus_vect"]["modif_subs_id"][0])] += 1


```

```python
actions_act_distr = defaultdict(int)

for key, val in actions.items():
    for i,act in enumerate(val):
        if act != {}: # edge case
            if i < len(val) - 1: # just to handle the edge case 
                if val[i+1] == act:
                # print("i am here")
                    actions_act_distr[-1] += 1
                else:
                    actions_act_distr[int(act["set_bus_vect"]["modif_subs_id"][0])] += 1
            else:
                # try: # handle occassional empty action
                actions_act_distr[int(act["set_bus_vect"]["modif_subs_id"][0])] += 1  
                # except: 
                #     print(act)
```
